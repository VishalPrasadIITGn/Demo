```python
x=np.linspace(0,2,1000)
fig,ax=plt.subplots(1,figsize=(6,6))
plt.title('graph')

def show(a,b,c,t):
    ax.clear()
    y=a*np.sin(2*np.pi*(c*x+b))
    ax.plot(x,y)
    plt.show()
a=wg.IntSlider( value=0,min=0,max=10,step=1,description='amp:')
b=wg.FloatSlider(value=0,min=0,max=10.0,step=0.1,description='phase:')
c=wg.FloatSlider(value=0,min=0,max=10.0,step=0.1,description='freq:') 
t=widgets.Text(value='sin',placeholder='Type something',description='String:')
wg.interactive(show,a=a,b=b,c=c,t=t)
```
